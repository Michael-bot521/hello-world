```python
import geoviews as gv
import geoviews.feature as gf
import xarray as xr
from cartopy import crs

gv.extension('bokeh')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
import seaborn as sns
regions = gpd.read_file("regions.geojson")
```

```python
df_covid = pd.read_csv("chiffres-cles.csv")
```

```python
#plt.figure(figsize=(12,12))
#sns.heatmap(df_covid.isna(), cbar = "p")
```

```python
df_regions = df_covid.loc[df_covid.granularite == "region"].groupby(by="maille_nom").sum()
```

```python
regions
```

```python
df_regions = df_regions.reset_index()
```

```python
df_regions = df_regions[['maille_nom', 'cas_confirmes','deces', 'reanimation']]
```

```python
df_regions = df_regions.rename(columns={"maille_nom": "nom"})
```

```python
geo_regions = regions.merge(df_regions, on = "nom")
```

```python
#geo_regions = geo_regions.drop("deces", axis = 1)
#geo_regions = geo_regions.drop("reanimation", axis = 1)
```

```python
geo_regions = geo_regions.loc[(geo_regions.nom != "Guadeloupe") & (geo_regions.nom != "Martinique") & (geo_regions.nom != "Guyane") & (geo_regions.nom != "La Runion") & (geo_regions.nom != "Mayotte")]
```

```python
geo_regions
```

```python
# cartographie statique
geo_regions.plot(column = "cas_confirmes", cmap='OrRd', legend= True, figsize = (14,14))
# plt.savefig("france_covid.png")
plt.show  # optionnel
```

```python
regions_covid = gv.Polygons(geo_regions, vdims=['nom','cas_confirmes'])
```

```python
from geoviews import dim
regions_covid.opts(width=800, height=600, toolbar='above', color=dim('cas_confirmes'), cmap='Blues',
          colorbar=True, tools=['hover'], aspect='equal')
```
